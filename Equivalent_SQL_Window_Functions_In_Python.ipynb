```python
# Importing necessary libraries
import pandas as pd
import datetime
import numpy as np
```

```python
# Importing csv file
df=pd.read_csv('Desktop/Sample_Orders.csv')
df
```

```python
# Copying data from one dataframe to another dataframe
df1=df[:]
df2=df[:]
df3=df[:]
df4=df[:]
df5=df[:]
df6=df[:]
df7=df[:]
df8=df[:]
df9=df[:]
df10=df[:]
df11=df[:]
```

# Running Sum (Cumulative Sum) based on order date

```python
# Sorting data based on order date (ascending)
df=df.sort_values(by='order_datetime',ascending=True).reset_index(drop=True)
df
```

```python
# Overall Running Sum
import itertools
df['running_sum']=pd.Series(itertools.accumulate(df['amount']))
df
```

```python
# Alternative solution for overall running sum
df['cumuluative_sum']=df['amount'].cumsum()
df
```

```python
# Customer id wise running sum of amount based on order date
df1['running_sum']=df1.groupby('customer_id')['amount'].cumsum()
df1.sort_values(by='customer_id',inplace=True)
df1
```

# Row number based on order date

```python
# Converting to timestamp (nanoseconds) and dividing by 10^9
df2['unixtime']=pd.to_datetime(df2['order_datetime']).astype(np.int64)/10**9

# Overall row number

# method='first' assigns rank in order they appear in the dataframe i.e. row number
df2['rownum']=df2['unixtime'].rank(method='first')
df2
```

```python
# Customer id wise row number based on order date

# Converting to timestamp (nanoseconds) and dividing by 10^9
df3['unixtime']=pd.to_datetime(df3['order_datetime']).astype(np.int64)/10**9

df3['rownum']=df3.groupby('customer_id')['unixtime'].rank(method='first')
df3.sort_values(by='customer_id',inplace=True)
df3
```

# Rank based on amount

```python
# Overall rank

# method='min' assigns rank
df4['rank']=df4['amount'].rank(method='min',ascending=False)
df4
```

```python
# customer id wise rank based on amount
df5['rank']=df5.groupby('customer_id')['amount'].rank(method='min',ascending=False)
df5.sort_values(by='customer_id',inplace=True)
df5
```

# Dense Rank based on amount

```python
# Overall dense rank

# method='dense' assigns dense rank
df6['dense_rank']=df6['amount'].rank(method='dense',ascending=False)
df6
```

```python
# customer id wise rank based on amount
df7['rank']=df7.groupby('customer_id')['amount'].rank(method='dense',ascending=False)
df7.sort_values(by='customer_id',inplace=True)
df7
```

# Amount value in the next row (lead)

```python
# amount value in the next row (lead)

# When offset value is -ve inside shift function, it represents lead
# -1 denotes next row
df8['next_amount']=df8['amount'].shift(-1)
df8
```

```python
# customer id wise amount value in the next row (lead)
df9['next_amount']=df9.groupby('customer_id')['amount'].shift(-1)
df9.sort_values(by='customer_id',inplace=True)
df9
```

# Amount value in the previous row (lag)

```python
# amount value in the previous row (lag)

#When offset value is +ve inside shift function, it represents lag
# +1 denotes previous row
df10['previous_amount']=df10['amount'].shift(1)
df10
```

```python
# customer id wise amount value in the previous row (lag)
df11['previous_amount']=df11.groupby('customer_id')['amount'].shift(1)
df11.sort_values(by='customer_id',inplace=True)
df11
```
